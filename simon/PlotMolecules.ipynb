```python
import scanpy as sc
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
import matplotlib.pyplot as plt
IPythonConsole.ipython_useSVG=False
```

```python
an = sc.read("datasets/trapnell_cpa_subset.h5ad")
```

```python
# Test in a kinase inhibitor
mol = Chem.MolFromSmiles("C1CC2=C3C(=CC=C2)C(=CN3C1)[C@H]4[C@@H](C(=O)NC4=O)C5=CNC6=CC=CC=C65")
# Default
mol
Draw.MolsToGridImage([mol, mol, mol], legends=["0","0","0"])
```

```python
s = [Chem.CanonSmiles(smiles) for smiles in an.obs["SMILES"].unique()]
s.sort(key=len)
s = s[4:7]
```

```python
img = Draw.MolsToGridImage([Chem.MolFromSmiles(smiles) for smiles in s], legends=s)
with open("../Latex/figures/03_technical_background/Trapnell_Molecules.png", "wb") as f:
    f.write(img.data)
img
```
